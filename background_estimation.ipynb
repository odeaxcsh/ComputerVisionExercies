## Computer Vision Course Material - Background Estimation

***Note:*** To run this notebook, you need to have the following packages installed:
+ `numpy`
+ `matplotlib`
+ `opneccv`

```python
import time
import cv2 as cv
import numpy as np
import matplotlib.animation as animation
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
vc = cv.VideoCapture('Data/videoplayback.mp4')
length = int(vc.get(cv.CAP_PROP_FRAME_COUNT))

images = np.array([vc.read()[1] for i in range(length)])
images = images[380:2090]
```

```python
image_size = images[0].shape

memory_size = np.prod(images.shape)

print('Memory size: {:.2f} MB ({} images)'.format(memory_size / 1024**2, length))
```

```python
start_time = time.time()
background = np.median(images, axis=0).astype(np.uint8)
end_time = time.time()

print('Took {:.2f} seconds to compute background'.format(end_time - start_time))
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(background)
plt.axis('off')
plt.show()
```
